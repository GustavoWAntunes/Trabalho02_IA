```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#a) Faça o download dos dados do arquivo data.mat ou data.csv. A primeira coluna é o tamanho da casa,
#a segunda coluna é o número de quartos, e a terceira coluna é o preço da casa.
df = pd.read_csv("/content/sample_data/data.csv", names=["Tamanho", "Quartos", "Preço"])
```

```python
#b) Utilize o comando python .describe()para fazer uma primeira análise estatística da sua base de dados.
# Qual a média de preço das casas? Quanto custa a menor casa? Quantos quartos tem a casa mais cara?
print(df.describe())

print(f"Média de preço: {df['Preço'].mean():.2f}") # puxa média do preço
print(f"Preço da menor casa: {df.loc[df['Tamanho'].idxmin(), 'Preço']}") # idxmin -> índice menor casa
print(f"Quartos da casa mais cara: {df.loc[df['Preço'].idxmax(), 'Quartos']}") # idmax -> índice maior casa
```

```python
#c) Gere uma matriz X para as variáveis ​​independentes (que são o tamanho da casa e o número de quartos) e o vetor y
# da variável dependente (que é o preço).

# gera a matriz com a matriz com as variáveis tamanho, quartos e preço, que mais tarde será usada para o cálculo da linha de regressão
X = df[['Tamanho', 'Quartos']]
y = df['Preço']
```

```python
# d) Verifique a correlação e a regressão para Tamanho da casa e Preço, e, Número de quartos e Preço e
# apresente os valores no gráfico de dispersão.

# Tamanho vc Preço
grafico, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(x=df['Tamanho'], y=df['Preço'], ax=axes[0])
modelo_tamanho = LinearRegression()
modelo_tamanho.fit(df[['Tamanho']], df['Preço'])
previsao_tamanho = modelo_tamanho.predict(df[['Tamanho']])
axes[0].plot(df['Tamanho'], previsao_tamanho, color='red') # linha de regressão
axes[0].set_title('Tamanho vs Preço')

# Quartos vs Preço
sns.scatterplot(x=df['Quartos'], y=df['Preço'], ax=axes[1])
modelo_quartos = LinearRegression()
modelo_quartos.fit(df[['Quartos']], df['Preço'])
previsao_quartos = modelo_quartos.predict(df[['Quartos']])
axes[1].plot(df['Quartos'], previsao_quartos, color='red') # linha de regressão
axes[1].set_title('Quartos vs Preço')

plt.show()
```

```python
# e) Calcule a Regressão Múltipla. Faça o gráfico de dispersão em 3D com o tamanho da casa, número de quartos, e o preço da casa.
# Neste caso iremos trabalhar com o espaço 3D (verifique como usar Axes3D).

modelo = LinearRegression().fit(X, y) # passando X e y da matriz que foram definidos anteriormente
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Tamanho'], df['Quartos'], df['Preço'], c='r')
ax.set_xlabel('Tamanho')
ax.set_ylabel('Quartos')
ax.set_zlabel('Preço')
plt.title('Dispersão 3D')
plt.show()
```

```python
# f) Trace a linha da regressão no Gráfico de Dispersão. Você pode girar este gráfico para visualizar melhor os dados.
xx, yy = np.meshgrid(np.linspace(df['Tamanho'].min(), df['Tamanho'].max(), 100), np.linspace(df['Quartos'].min(), df['Quartos'].max(), 100))
zz = modelo.intercept_ + modelo.coef_[0] * xx + modelo.coef_[1] * yy # equação linear multiplra
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Tamanho'], df['Quartos'], df['Preço'], c='r')
ax.plot_surface(xx, yy, zz, color='blue', alpha=0.5)
plt.title('Regressão Múltipla 3D')
ax.set_xlabel('Tamanho')
ax.set_ylabel('Quartos')
ax.set_zlabel('Preço')
plt.show()
```

```python
# g) Mostre na figura os coeficientes de correlação entre Tamanho da casa e Preço, e Número de quartos e Preço.

correlacao_tamanho_preco = df['Tamanho'].corr(df['Preço'])
correlacao_quartos_preco = df['Quartos'].corr(df['Preço'])
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# Tamanho vs Preço
sns.scatterplot(x=df['Tamanho'], y=df['Preço'], ax=axes[0])
axes[0].set_title(f'Tamanho vs Preço\nCorrelação: {correlacao_tamanho_preco:.2f}')
modelo_tamanho = LinearRegression()
modelo_tamanho.fit(df[['Tamanho']], df['Preço'])
previsao_tamanho = modelo_tamanho.predict(df[['Tamanho']])
axes[0].plot(df['Tamanho'], previsao_tamanho, color='red')
# Quartos vs Preço
sns.scatterplot(x=df['Quartos'], y=df['Preço'], ax=axes[1])
axes[1].set_title(f'Quartos vs Preço\nCorrelação: {correlacao_quartos_preco:.2f}')
modelo_quartos = LinearRegression()
modelo_quartos.fit(df[['Quartos']], df['Preço'])
previsao_quartos = modelo_quartos.predict(df[['Quartos']])
axes[1].plot(df['Quartos'], previsao_quartos, color='red')
plt.tight_layout()
plt.show()
```
